```python
# Function to classify an individual as a user (1) or non-user (0) of a drug
# 0- Never used or used the drug over a decade ago
# 1- Used the drug in month, week or day
def change_drug(x):
    
    if ((x == 'CL0') or (x == 'CL1') ):
        x = 0
    
    elif ((x == 'CL2') or (x == 'CL3') or (x == 'CL6') or (x == 'CL5') or (x == 'CL4') ):
        x = 1
    
    return x
```

```python
# Applying function to all drugs
data['Alcohol'] = data['Alcohol'].map(change_drug)
data['Amphet'] = data['Amphet'].map(change_drug)
data['Amyl'] = data['Amyl'].map(change_drug)
data['Benzos'] = data['Benzos'].map(change_drug)
data['Caff'] = data['Caff'].map(change_drug)
data['Cannabis'] = data['Cannabis'].map(change_drug)
data['Choc'] = data['Choc'].map(change_drug)
data['Coke'] = data['Coke'].map(change_drug)
data['Crack'] = data['Crack'].map(change_drug)
data['Ecstasy'] = data['Ecstasy'].map(change_drug)
data['Heroin'] = data['Heroin'].map(change_drug)
data['Ketamine'] = data['Ketamine'].map(change_drug)
data['Legalh'] = data['Legalh'].map(change_drug)
data['LSD'] = data['LSD'].map(change_drug)
data['Meth'] = data['Meth'].map(change_drug)
data['Mushrooms'] = data['Mushrooms'].map(change_drug)
data['Nicotine'] = data['Nicotine'].map(change_drug)
data['Semer'] = data['Semer'].map(change_drug)
data['VSA'] = data['VSA'].map(change_drug)
```

```python
#Decoding Age
#0-'18-24'age 
#1-'25-34'age 
#2-'35-44'age 
#3-'45-54'age 
#4-'55-64'age 
#5-'65+'age
def change_age(x):
    if (x == -0.95197):
        x = 0
    elif (x == -0.07854):
        x = 1
    elif (x == 0.49788):
        x = 2
    elif (x == 1.09449):
        x = 3
    elif (x == 1.82213):
        x = 4
    elif (x == 2.59171):
        x = 5
    return x

data['Age'] = data['Age'].map(change_age)
```

```python
#Decoding Gender
#0- Female
#1- Male 
def change_gender(x):
    if (x == 0.48246 ):
        x = 0
    elif (x == -0.48246 ):
        x = 1
    return x

data['Gender'] = data['Gender'].map(change_gender)
```

```python
#Decoding Education
#0- Left school before 16 years                          
#1- Left school at 16 years                              
#2- Left school at 17 years                             
#3- Left school at 18 years                              
#4- Some college or university, no certificate or degree 
#5- Professional certificate/ diploma                    
#6- University degree                                   
#7- Masters degree                                     
#8- Doctorate degree                    
def change_education(x):
    if (x == -2.43591):
        x = 0
    elif (x == -1.73790):
        x = 1
    elif (x == -1.43719):
        x = 2
    elif (x == -1.22751):
        x = 3
    elif (x == -0.61113):
        x = 4
    elif (x == -0.05921):
        x = 5
    elif (x == 0.45468):
        x = 6
    elif (x == 1.16365):
        x = 7
    elif (x == 1.98437):
        x = 8
    return x

data['Education'] = data['Education'].map(change_education)
```

```python
#Decoding country
#0- Australia 
#1- Canada   
#2- New Zealand
#3- Other     
#4- Republic of Ireland 
#5- UK       
#6- USA
def change_country(x):
    if (x == -0.09765):
        x = 0
    elif (x == 0.24923):
        x = 1
    elif (x == -0.46841):
        x = 2
    elif (x == -0.28519):
        x = 3
    elif (x == 0.21128):
        x = 4
    elif (x == 0.96082):
        x = 5
    elif (x == -0.57009):
        x = 6
    return x

data['Country'] = data['Country'].map(change_country)
```

```python
#Decoding Ethnicity
#0- Asian  
#1-Black  
#2- Mixed-Black/Asian 
#3- Mixed-White/Asian 
#4- Mixed-White/Black 
#5- Other           
#6- White             
def change_ethnicity(x):
    if (x == -0.50212):
        x = 0
    elif (x == -1.10702):
        x = 1
    elif (x == 1.90725):
        x = 2
    elif (x == 0.12600):
        x = 3
    elif (x == -0.22166):
        x = 4
    elif (x == 0.11440):
        x = 5
    elif (x == -0.31685):
        x = 6
    return x

data['Ethnicity'] = data['Ethnicity'].map(change_ethnicity)
```

```python
#Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve, precision_score, recall_score
```

```python
def preprocessing(data, column):
    data = data.copy()
    
    # Splitting the data into X and y
    y = data[column]
    X = data.drop(column, axis=1)
    
    #Spliting the data on target dataset into train and test using train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
    
    # Scale X
    scaler = StandardScaler()
    scaler.fit(X_train)
    
    X_train = pd.DataFrame(scaler.transform(X_train), 
                           index=X_train.index, 
                           columns=X_train.columns)
    X_test = pd.DataFrame(scaler.transform(X_test), 
                          index=X_test.index, 
                          columns=X_test.columns)
    
    return X_train, X_test, y_train, y_test
```

```python
# Different models  
classifiers = {'DecisionTreeClassifier': DecisionTreeClassifier(),
          'Random Forest Classifier': RandomForestClassifier(),
          'Support Vector Machines': SVC(probability=True),
          'KNeighborsClassifier' : KNeighborsClassifier()}
```

```python
# Data Preprocessing for Nicotine
X_train, X_test, y_train, y_test = preprocessing(Nicotine_df, 'Nicotine_User')

#Printing size of training and testing dataset 
print('Trai`n set:', X_train.shape, y_train.shape)
print('Test set:', X_test.shape, y_test.shape)
```

```python
# Function to Evaluate model on testing dataset
def model_evaluation(y_test, y_preds):
    
    acc = accuracy_score(y_test, y_preds)
    print(key + ' Accuracy: {:.2%}'.format(acc))
    
    p = precision_score(y_test, y_preds)
    print(key + ' precision-Score: {:.5}'.format(p))
    
    r = recall_score(y_test, y_preds)
    print(key + ' recall-Score: {:.5}'.format(r))
```
